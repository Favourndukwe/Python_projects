```python
mean_created_date = df['created_date'].mean().round()
mean_created_date
```

```python
df['CREATED_DATE'] = df['created_date'].fillna(mean_created_date)
df
```

```python
df_dropped = df.drop(['category', 'Country','created_date','Abbreviation','channel_type','VIDEO_VIEWS_RANK'], axis=1)
df_dropped
```

```python
mean_video_views_rank= pd.DataFrame(df.groupby('CATEGORY')['video_views_rank'].mean())
mean_video_views_rank
```

```python
category = df.category.unique().tolist()
for i in category:
        if df['category'].isna().any():
            if i in mean_video_views_rank:
                df.loc[df['category'] == i, 'video_views_rank'] = df.loc[df['category'] == i, 'video_views_rank'].fillna(mean_video_views_rank[i])            
```
